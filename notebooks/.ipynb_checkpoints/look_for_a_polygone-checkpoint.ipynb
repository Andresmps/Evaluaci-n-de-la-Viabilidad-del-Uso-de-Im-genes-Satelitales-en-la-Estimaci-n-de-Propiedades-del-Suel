```python
path_municipios = os.path.abspath('../data/external/geojsons/colombia_municipios.geojson')
municipios = gpd.read_file(path_municipios)

municipios['MPIO_CNMBR'] = municipios['MPIO_CNMBR'].apply(lambda x: unidecode(x))
municipios['DPTO_CNMBR'] = municipios['DPTO_CNMBR'].apply(lambda x: unidecode(x))
print(municipios.shape)
municipios.head()
```

```python
path_departamentos = os.path.abspath('../data/external/geojsons/colombia_departamentos.geojson')
departamentos = gpd.read_file(path_departamentos)
print(departamentos.shape)
departamentos.head()
```

```python
path_nirs = os.path.abspath(
    "../data/raw/\soil_cane_vis_NIRS/soil_cane_vis_NIR_georeferenced.csv"
)

nir_samples = pd.read_csv(path_nirs, sep=';')
nir_samples.rename(columns={'Longitud.x':'lon'}, inplace=True)
nir_samples.rename(columns={'Latitud.x':'lat'}, inplace=True)
nir_samples['lon'] = nir_samples['lon'].apply(lambda x:x.replace(',', '.')).astype(float)
nir_samples['lat'] = nir_samples['lat'].apply(lambda x:x.replace(',', '.')).astype(float)
nir_samples.drop(['i'], axis=1, inplace=True)

print(nir_samples.shape)
nir_samples.head()
```

```python
nir_samples_gdf = gpd.GeoDataFrame(
    nir_samples, geometry=gpd.points_from_xy(nir_samples['lon'], nir_samples['lat'])
)
nir_samples_gdf.head()
```

```python
ax1 = departamentos[
    departamentos['NOMBRE_DPT'].isin(['BOYACA', 'SANTANDER'])
].plot(color='white', edgecolor='black')

nir_samples_gdf.plot(ax=ax1, color='red')

plt.show()
```

```python
ax2 = municipios[
    municipios['DPTO_CNMBR'].isin(['BOYACA', 'SANTANDER'])
].plot(color='white', edgecolor='black')

nir_samples_gdf.plot(ax=ax2, color='red')

plt.show()
```
