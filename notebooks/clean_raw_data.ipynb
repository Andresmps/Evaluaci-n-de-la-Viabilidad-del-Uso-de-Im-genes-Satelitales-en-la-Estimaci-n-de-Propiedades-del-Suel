```python
def check_path(path):
    files = path.split('/')
    tmp_path = ''
    
    for file in files:
        tmp_path += f"{file}/"
        if not os.path.exists(tmp_path):
            os.mkdir(tmp_path)
        else:
            print(f'File: **{file}** exists')
```

#### soil_cane_vis_NIR_georeferenced

```python
df_soil_cane_vis_NIR_georeferenced = pd.read_csv(
    '../data/raw/soil_cane_vis_NIRS/soil_cane_vis_NIR_georeferenced.csv', sep=';'
)
print(df_soil_cane_vis_NIR_georeferenced.shape)
df_soil_cane_vis_NIR_georeferenced.head()
```

```python
df_soil_cane_vis_NIR_georeferenced_ = df_soil_cane_vis_NIR_georeferenced.copy()
```

```python
df_soil_cane_vis_NIR_georeferenced_ = df_soil_cane_vis_NIR_georeferenced_[
    df_soil_cane_vis_NIR_georeferenced_['inconsistencias'] == 0
]
```

```python
spectrum_cols = df_soil_cane_vis_NIR_georeferenced_.columns.tolist()[37:]
```

```python
cols_ = [
    'Longitud.x', 'Latitud.x', 'Departamento', 'Municipio',
    'Cultivo', 'Topografia', 'pH', 'MO.x', 'Ca', 'Mg', 'K', 'Na'
]
cols_.extend(spectrum_cols)

df_soil_cane_vis_NIR_georeferenced_ = df_soil_cane_vis_NIR_georeferenced_[cols_]
```

```python
df_soil_cane_vis_NIR_georeferenced_ = df_soil_cane_vis_NIR_georeferenced_.rename(
    columns={
        'Longitud.x': 'lon',
        'Latitud.x': 'lat',
        'MO.x': 'OM',
        'Departamento': 'departamento',
        'Municipio': 'municipio',
        'Cultivo': 'cultivo',
        'Topografia': 'topografia'
    }
)
```

```python
float_cols = spectrum_cols.copy()
float_cols.extend(
    ['lat', 'lon', 'pH', 'OM', 'Ca', 'Mg', 'K', 'Na']
)
print(float_cols[:5])
```

```python
for col in float_cols:
    df_soil_cane_vis_NIR_georeferenced_[col] = df_soil_cane_vis_NIR_georeferenced_[col].apply(
        lambda x: x.replace(',', '.')
    ).astype(float)
```

```python
cols = df_soil_cane_vis_NIR_georeferenced_.columns
```

```python
df_soil_cane_vis_NIR_georeferenced_.columns = [i.replace(',', '.') for i in cols]
```

```python
df_soil_cane_vis_NIR_georeferenced_.info()
```

```python
print(df_soil_cane_vis_NIR_georeferenced_.shape)
df_soil_cane_vis_NIR_georeferenced_.head()
```

```python
df_soil_cane_vis_NIR_georeferenced_.to_csv(
    '../data/processed_raw/soil_cane_vis_NIRS/soil_cane_vis_NIR_georeferenced.csv',
    sep=';',
    decimal='.',
    index=False
)
```

```python
df_remuestreo_2020_muestra_tejido_.head(2)
```

```python
float_cols = ['lat', 'lon', 'pH', 'OM', 'Ca', 'Mg', 'K', 'Na']

for col in float_cols:
    print(col)
    df_remuestreo_2020_muestra_tejido_[col] = df_remuestreo_2020_muestra_tejido_[col].apply(
        lambda x: str(x).replace(r'<', '').replace(r'Z', '').replace(',', '.')
    ).astype(float)
```
