```python
import pandas as pd
import os

from IPython.display import display 
```

```python
pd.options.display.max_rows = 999
pd.options.display.max_colwidth = 999
pd.options.display.max_columns = 999
pd.options.display.float_format = '{:.3f}'.format
```

```python
properties = [
    'pH',
    'OM',
    'Ca',
    'Mg',
    'K',
    'Na'
]

models = [
    'LR',
    'SVR',
    'LASSO',
    'PLSR2',
    'PLSR6',
    'RF'
]

scalers = [
#     'wo_norm',
    'minmax',
    'standard'
]

satellites = [
    'landsat_simulated',
    'landsat_simulated_indices',
    'sentinel2_simulated',
    'sentinel2_simulated_indices',
    'desis_simulated',
    'sentinel2_soil',
    'sentinel2_soil_indices',
    'sentinel2_all',
    'sentinel2_all_indices',
    'vis_nir'
]

transformations = [
    'scE_d1_d2_',
    'scE_d1_d2_fft_',
    'scE_d1_d2_cr_',
    'scE_d1_d2_fft_cr_'
]
```

```python
all_satelital_metrics_data = []

for satellite in satellites:
    for scaler in scalers:
        for transformation in transformations:
            for property_ in properties:
                path = (
                    f'../data/models/predictions/{satellite}/{scaler}/' +
                    f'{transformation}/{property_}/metrics.csv'
                )
                df_metrics = pd.read_csv(path, sep=';', decimal='.')
                all_satelital_metrics_data.append(df_metrics)
                
#                 temp = []
#                 for path in files_paths:
#                     metrics_path_ = os.path.abspath(f'../data/satellite_data/{satellite}/results/{path}') 
#                     temp.append(pd.read_csv(metrics_path_, sep=';', decimal='.'))
#                 temp_ = pd.concat(temp)
#                 
```

```python
all_metrics = pd.concat(all_satelital_metrics_data)
```

```python
def check_path(path):
    files = path.split('/')
    tmp_path = ''
    
    for file in files:
        tmp_path += f"{file}/"
        if not os.path.exists(tmp_path):
            os.mkdir(tmp_path)
        else:
            print(f'File: **{file}** exists')
```

### Creating tables

```python
for property_ in list(all_metrics['Property'].unique()):
    for scaler in list(all_metrics['Scaler'].unique()):
        for features in list(all_metrics['Features'].unique()):
            temp_df = all_metrics[
                (all_metrics['Property'] == property_) &
                (all_metrics['Scaler'] == scaler) &
                (all_metrics['Features'] == features)
            ].sort_values(
                by=[
                    'R2 CV All', 'R2 Test All'
                ],
                ascending=False
            ).copy().reset_index(drop=True)
            test_data_path_ = str(
                "../reports/metrics/" +
                f"{scaler}/{features}"
            )
            check_path(test_data_path_)
                
            temp_df.to_csv(
                f"{test_data_path_}/result_metrics_{property_}.csv",
                sep=';',
                decimal='.',
                index=False
            )
            display(temp_df.head(10))
```
