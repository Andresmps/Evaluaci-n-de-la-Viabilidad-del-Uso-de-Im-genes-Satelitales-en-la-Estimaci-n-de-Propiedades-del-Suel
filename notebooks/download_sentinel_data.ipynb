```python
import pandas as pd

from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from collections import OrderedDict
from datetime import date
```

```python
pd.options.display.max_rows = 999
pd.options.display.max_colwidth = 999
pd.options.display.max_columns = 999
```

```python
# connect to the API
api = SentinelAPI('andresmartinez', '123456789', 'https://scihub.copernicus.eu/dhus')
api
```

```python
area = "POLYGON ((-73.7 5.85, -73.3 5.85, -73.3 6.2, -73.7 6.2, -73.7 5.85))"

query_kwargs = {
    'platformname': 'Sentinel-2',
    'producttype': 'S2MSI1C',
    'date': (date(2020, 3, 1), date(2020, 3, 30)),
    'area': area
}

products = OrderedDict()
pp = api.query(**query_kwargs)
products.update(pp)

products_df = api.to_dataframe(products)

products_df_sorted = products_df.sort_values(['cloudcoverpercentage', 'ingestiondate'], ascending=[True, True])
print(products_df_sorted.shape)
products_df_sorted
```
